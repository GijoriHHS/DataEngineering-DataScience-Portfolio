```python
# import pandas as pd
# import sqlite3 as sq

# go_sales_conn = sq.connect("go_sales_train.sqlite")
# go_crm_conn = sq.connect("go_crm_train.sqlite")
# go_staff_conn = sq.connect("go_staff_train.sqlite")

# import pyodbc

# DB = {
#     'servername': r'LAPTOP-VAN-GIJO\SQLEXPRESS',
#     'database': 'Source Data Model'
# }

# export_conn = pyodbc.connect(
#     'DRIVER={SQL Server};SERVER=' + 
#     DB['servername'] + ';DATABASE=' + 
#     DB['database'] + ';Trusted_Connection=yes'
#     )
# export_cursor = export_conn.cursor()
# export_conn

# export_cursor.execute("DELETE FROM sales_territory")
# export_conn.commit()


# df = pd.read_sql("SELECT * FROM sales_territory", go_crm_conn )
# df

# for index, row in df.iterrows():
#     try:
#         query = f"INSERT INTO sales_territory VALUES ({row['SALES_TERRITORY_CODE']}, '{row['TERRITORY_NAME_EN']}')"
#         export_cursor.execute(query)
#     except pyodbc.Error:
#         print(query)

# export_conn.commit()
# export_cursor.close()

#-------------------------------------------------------------------------------------------------------------------

# import pandas as pd
# import sqlite3 as sq
# import pyodbc

# go_sales_conn = sq.connect("go_sales_train.sqlite")
# go_crm_conn = sq.connect("go_crm_train.sqlite")
# go_staff_conn = sq.connect("go_staff_train.sqlite")

# DB = {
#     'servername': r'LAPTOP-VAN-GIJO\SQLEXPRESS',
#     'database': 'Source Data Model'
# }

# export_conn = pyodbc.connect(
#     'DRIVER={SQL Server};SERVER=' + 
#     DB['servername'] + ';DATABASE=' + 
#     DB['database'] + ';Trusted_Connection=yes'
# )
# export_cursor = export_conn.cursor()

# tabellen = [
#     ("sales_territory", ["SALES_TERRITORY_CODE", "TERRITORY_NAME_EN"]),
#     ("age_group", ["AGE_GROUP_CODE", "UPPER_AGE", "LOWER_AGE"]),
#     ("retailer_segment", ["SEGMENT_CODE", "LANGUAGE", "SEGMENT_NAME", "SEGMENT_DESCRIPTION"]),
#     ("retailer_type", ["RETAILER_TYPE_CODE", "RETAILER_TYPE_EN"]),
    
#     ("country", ["COUNTRY_CODE", "COUNTRY_EN", "FLAG_IMAGE", "SALES_TERRITORY_CODE"]),
#     ("retailer_headquarters", ["RETAILER_CODEMR", "RETAILER_NAME", "ADDRESS1", "ADDRESS2", "CITY", "REGION", "POSTAL_ZONE", "COUNTRY_CODE", "PHONE", "FAX", "SEGMENT_CODE"]),

#     ("retailer", ["RETAILER_CODE", "RETAILER_CODEMR", "COMPANY_NAME", "RETAILER_TYPE_CODE"]),
#     ("retailer_site", ["RETAILER_SITE_CODE", "RETAILER_CODE", "ADDRESS1", "ADDRESS2", "CITY", "REGION", "COUNTRY_CODE", "ACTIVE_INDICATOR"]),
#     ("sales_demographic", ["DEMOGRAPHIC_CODE", "RETAILER_CODEMR", "AGE_GROUP_CODE", "SALES_PERCENT"]),
#     ("retailer_contact", ["RETAILER_CONTACT_CODE", "RETAILER_SITE_CODE", "FIRST_NAME", "LAST_NAME", "JOB_POSITION_EN", "EXTENSION", "FAX", "E_MAIL", "GENDER"]),
# ]

# def clear_table(tabel_naam):

#     try:
#         export_cursor.execute(f"DELETE FROM {tabel_naam}")
#         export_conn.commit()
#         print(f"Data verwijderd uit {tabel_naam}")
#     except pyodbc.Error as e:
#         print(f"Fout bij verwijderen van {tabel_naam}: {e}")

# def transfer_table(sqlite_conn, tabel_naam, columns):
#     query = f"SELECT * FROM {tabel_naam}"
#     df = pd.read_sql(query, sqlite_conn)
#     # print(df)
#     df = df.where(pd.notnull(df), None)  
#     # print(f"nieuwe df {df}")
#     placeholders = ", ".join(["?" for _ in columns])  
#     insert_query = f"INSERT INTO {tabel_naam} ({', '.join(columns)}) VALUES ({placeholders})"

#     for row in df.itertuples(index=False):
#         try:
#             export_cursor.execute(insert_query, row)
#         except pyodbc.Error as e:
#             print(f"Error bij invoegen in {tabel_naam}: {row}\nFoutmelding: {e}")

#     export_conn.commit()
#     print(f"Data overgezet naar {tabel_naam}")


# for tabel_naam, _ in reversed(tabellen): 
#     clear_table(tabel_naam)

# for tabel_naam, columns in tabellen:
#     transfer_table(go_crm_conn, tabel_naam, columns)

# export_cursor.close()
# export_conn.close()
# go_sales_conn.close()
# go_crm_conn.close()
# go_staff_conn.close()

# -----------------------------------------------------------------------------------------

import pandas as pd
import sqlite3 as sq
import pyodbc

# Verbinding maken met SQLite-databases
go_sales_conn = sq.connect("go_sales_train.sqlite")
go_crm_conn = sq.connect("go_crm_train.sqlite")
go_staff_conn = sq.connect("go_staff_train.sqlite")

# SQL Server-verbindingsdetails
DB = {
    'servername': r'LAPTOP-VAN-GIJO\SQLEXPRESS',
    'database': 'Source Data Model'
}

# Verbinding maken met SQL Server
export_conn = pyodbc.connect(
    'DRIVER={SQL Server};SERVER=' + 
    DB['servername'] + ';DATABASE=' + 
    DB['database'] + ';Trusted_Connection=yes'
)
export_cursor = export_conn.cursor()

tabellen_crm_train = [
    ("sales_territory", ["SALES_TERRITORY_CODE", "TERRITORY_NAME_EN"]),
    ("age_group", ["AGE_GROUP_CODE", "UPPER_AGE", "LOWER_AGE"]),
    ("retailer_segment", ["SEGMENT_CODE", "LANGUAGE", "SEGMENT_NAME", "SEGMENT_DESCRIPTION"]),
    ("retailer_type", ["RETAILER_TYPE_CODE", "RETAILER_TYPE_EN"]),
    
    ("country", ["COUNTRY_CODE", "COUNTRY_EN", "FLAG_IMAGE", "SALES_TERRITORY_CODE"]),
    ("retailer_headquarters", ["RETAILER_CODEMR", "RETAILER_NAME", "ADDRESS1", "ADDRESS2", "CITY", "REGION", "POSTAL_ZONE" , "COUNTRY_CODE", "PHONE", "FAX", "SEGMENT_CODE"]),

    ("retailer", ["RETAILER_CODE", "RETAILER_CODEMR", "COMPANY_NAME", "RETAILER_TYPE_CODE"]),
    ("retailer_site", ["RETAILER_SITE_CODE", "RETAILER_CODE", "ADDRESS1", "ADDRESS2", "CITY", "REGION", "POSTAL_ZONE", "COUNTRY_CODE", "ACTIVE_INDICATOR"]),
    ("sales_demographic", ["DEMOGRAPHIC_CODE", "RETAILER_CODEMR", "AGE_GROUP_CODE", "SALES_PERCENT"]),
    ("retailer_contact", ["RETAILER_CONTACT_CODE", "RETAILER_SITE_CODE", "FIRST_NAME", "LAST_NAME", "JOB_POSITION_EN", "EXTENSION", "FAX", "E_MAIL", "GENDER"]),
]

tabel_staff = [
    ("course", ["COURSE_CODE", "COURSE_DESCRIPTION"]),
    ("sales_branch", ["SALES_BRANCH_CODE", "ADDRESS1", "ADDRESS2", "CITY", "REGION", "COUNTRY_CODE"]),
    ("satisfaction_type", ["SATISFACTION_TYPE_CODE", "SATISFACTION_TYPE_DESCRIPTION"]),
    ("sales_staff", ["SALES_STAFF_CODE", "FIRST_NAME", "LAST_NAME", "POSITION_EN", "WORK_PHONE", "EXTENSION", "FAX", "EMAIL", "DATE_HIRED", "SALES_BRANCH_CODE", "MANAGER_CODE"]),
    ("satisfaction", ["YEAR", "SALES_STAFF_CODE", "SATISFACTION_TYPE_CODE"]),
    ("training", ["YEAR", "SALES_STAFF_CODE", "COURSE_CODE"]),
]


# tabellen_crm_sales = [
#     ("order_method", ["ORDER_METHOD_CODE", "ORDER_METHOD_EN"]),
#     ("product_line", ["PRODUCT_LINE_CODE", "PRODUCT_LINE_EN"]),
#     ("country", ["COUNTRY_CODE", "COUNTRY", "LANGUAGE", "CURRENCY_NAME"]),

#     ("product_type", ["PRODUCT_TYPE_CODE", "PRODUCT_LINE_CODE", "PRODUCT_TYPE_EN"]),

#     ("sales_branch", ["SALES_BRANCH_CODE", "ADDRESS1", "ADDRESS2", "CITY", "REGION", "COUNTRY_CODE"]),
#     ("sales_staff", ["SALES_STAFF_CODE", "FIRST_NAME", "LAST_NAME", "POSITION_EN", "WORK_PHONE", "EXTENSION", "FAX", "EMAIL", "DATE_HIRED", "SALES_BRANCH_CODE"]),

#     ("retailer_site", ["RETAILER_SITE_CODE", "RETAILER_CODE", "ADDRESS1", "ADDRESS2", "CITY", "REGION", "COUNTRY_CODE", "ACTIVE_INDICATOR"]),

#     ("order_header", ["ORDER_NUMBER", "RETAILER_NAME", "RETAILER_SITE_CODE", "RETAILER_CONTACT_CODE", "SALES_STAFF_CODE", "SALES_BRANCH_CODE", "ORDER_DATE", "ORDER_METHOD_CODE"]),
    
#     ("product", ["PRODUCT_NUMBER", "INTRODUCTION_DATE", "PRODUCT_TYPE_CODE", "PRODUCTION_COST", "MARGIN", "PRODUCT_IMAGE", "LANGUAGE", "PRODUCT_NAME", "DESCRIPTION"]),

#     ("return_reason", ["RETURN_REASON_CODE", "RETURN_DESCRIPTION_EN"]),

#     ("returned_item", ["RETURN_CODE", "RETURN_DATE", "ORDER_DETAIL_CODE", "RETURN_REASON_CODE", "RETURN_QUANTITY"]),

#     ("order_details", ["ORDER_DETAIL_CODE", "ORDER_NUMBER", "PRODUCT_NUMBER", "QUANTITY", "UNIT_COST", "UNIT_PRICE", "UNIT_SALE_PRICE"]),
# ]


def clear_table(tabel_naam):
    try:
        export_cursor.execute(f"DELETE FROM {tabel_naam}")
        export_conn.commit()
        print(f"Data verwijderd uit {tabel_naam}")
    except pyodbc.Error as e:
        print(f"Fout bij verwijderen van {tabel_naam}: {e}")

def transfer_table(sqlite_conn, tabel_naam, columns):
    query = f"SELECT * FROM {tabel_naam}"
    df = pd.read_sql(query, sqlite_conn)

    df = df.where(pd.notnull(df), None)

    placeholders = ", ".join(["?" for _ in columns])
    insert_query = f"INSERT INTO {tabel_naam} ({', '.join(columns)}) VALUES ({placeholders})"

    for row in df.itertuples(index=False, name=None):
        try:
            row = [None if pd.isna(value) else value for value in row]
            # print(f"Inserting into {tabel_naam}: {row}")  # Print de rij om te controleren
            export_cursor.execute(insert_query, row)
        except pyodbc.Error as e:
            print(f"Error bij invoegen in {tabel_naam}: {row}\nFoutmelding: {e}")

    export_conn.commit()
    print(f"Data overgezet naar {tabel_naam}")

for tabel_naam, _ in reversed(tabellen_crm_train): 
    clear_table(tabel_naam)

for tabel_naam, columns in tabellen_crm_train:
    transfer_table(go_crm_conn, tabel_naam, columns)
    
for tabel_naam, _ in reversed(tabel_staff): 
    clear_table(tabel_naam)

for tabel_naam, columns in tabel_staff:
    transfer_table(go_staff_conn, tabel_naam, columns)
    
# for tabel_naam, _ in reversed(tabellen_crm_sales): 
#     clear_table(tabel_naam)

# for tabel_naam, columns in tabellen_crm_sales:
#     transfer_table(go_sales_conn, tabel_naam, columns)

export_cursor.close()
export_conn.close()
go_sales_conn.close()
go_crm_conn.close()
go_staff_conn.close()

# -----------------------------------------------------------------------------------------







```
